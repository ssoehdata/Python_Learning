## The hypothesis to be tested is that the  four categories of 
## Age, Sex, PassengerClass (PClass), and if the passenger survived (Survived) 
## can be used to predict the odds for survival

####  Some general cleaning to remove columns that will not be used in the analysis and printing the first 10 rows 

```python
df = df.drop(['Name','Siblings/Spouses Aboard', 'Parents/Children Aboard', 'Fare'], axis= 1)
df.head(10)
```

### Resolving the number of ages in the data into a more manageable number of groups. This will be done by creating a smaller subset of ages grouped into "buckets". These will be created using the pondas.cut() function.

```python
age_range_bins = [0,5, 10, 15, 20, 25, 30, 35, 40, 45, 50, 55, 60, 65, 70, 75,80]
```

```python
age_range_labels = [5, 10, 15, 20, 25, 30, 35, 40, 45, 50, 55, 60, 65, 70, 75,80]
```

```python
df['Age_Range']=pd.cut(df['Age'],age_range_bins,labels=age_range_labels)
```

### The created Age_Range column will be used in place of the Age column, so that can be dropped:

```python
df = df.drop(['Age'], axis=1)
```
